```python
import pandas as pd
import numpy as np
import re
import nltk
#nltk.download()
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.util import ngrams
from nltk import ngrams, FreqDist
from nltk.lm import NgramCounter
import string
```

# Transforming the Dataset

```python
# delete first row and column
records_996 = pd.read_csv("records_996.csv", header = None)
records_996 = records_996.drop_duplicates(keep=False, inplace=False)
records_996 = records_996.drop(records_996[records_996[2] == "[]"].index)
records_996 = records_996.replace(" <br/>", np.NaN)
records_996 = records_996.fillna('empty')
records_996 = records_996.replace('<.*?>','',regex=True)
records_996.to_csv('records_996_cleaned.csv')
```

```python
records_996
```

# Prepare the Dataset for NLTK Analysis

```python
records_996_string = records_996.copy()
records_996_string = records_996_string.drop(0, axis = 1)
records_996_string = records_996_string.drop(1, axis = 1)
records_996_string = records_996_string.values.tolist()
records_996_string = " ".join(str(v) for v in records_996_string)
records_996_string = re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", records_996_string)
records_996_string = records_996_string.replace("\'empty\'", "")
records_996_string = records_996_string.translate(str.maketrans('','',string.punctuation))
records_996_string
```

# NLTK Analysis

```python
records_996_tokenized = word_tokenize(records_996_string)
```

```python
stop_words = set(stopwords.words('english'))
records_996_filtered = [w for w in records_996_tokenized if not w in stop_words]

records_996_filtered = []

for w in records_996_tokenized:
    if w not in stop_words:
        records_996_filtered.append(w)

records_996_filtered = [w.lower() for w in records_996_filtered]

#records_996_filtered = [word for word in records_996_filtered if len(word) > 1]
```
